# 1.	Is it possible that an event is independent of itself? If so, when?

Possible case:
If A  and B are 2 independent events then P(A∩B)=P(A)P(B)
similarly if A is an event It is independent of itself iff P(A∩A)=P(A)P(A)=[P(A)]^2
This is possible only if the probability of the given event is either 0 or 1 as 0^2=0 and 1^2=1
